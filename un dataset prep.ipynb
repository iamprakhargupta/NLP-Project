```python
df = pd.read_excel(path_to_speakers_by_session, dtype=object, skip_blank_lines=True)
df.dropna(how='all', inplace=True)
df['Year'] = df[df.columns[0]]
df['year_session_iso'] = df['Year'].astype(str) + "_" + df['Session'].astype(str) + "_" + df['ISO Code'].astype(str)
df.index = df['year_session_iso']
df.drop(['year_session_iso'], 1, inplace=True)
df.head()
```

```python
speech_file_names = glob.glob(path_to_Converted_sessions+"/*/*.txt")
```

```python
speech_file_names
```

```python
def buildSet():
    for full_path in speech_file_names:
        filename = full_path.split('\\')[-1]
        parts = filename.split('_')
        country = parts[0]
        session = parts[1]
        year = parts[2][0:4]
        key = year + "_" + session + "_" + country
        doc = getDoc(full_path)
        yield dict(year_session_iso=key, doc=doc)
        
def getDoc(p):
    return open(p, 'r',encoding='utf-8').read()
```

```python
# pd.DataFrame([list(buildSet())[0]])
```

```python
doc_df = pd.DataFrame(list(buildSet()))
doc_df.index = doc_df['year_session_iso']
doc_df = doc_df[['doc']]
doc_df.head()
```

```python
joined = doc_df.join(df)
```
