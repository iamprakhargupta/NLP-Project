This code is for tokenisation and part of preprocessing step 


```python
import pandas as pd
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
df=pd.read_json("un_general_assembly.json",lines=True)
```

```python
df.head()
```

```python
def wordsplitter(x):
    return word_tokenize(x)
def sentsplitter(x):
    return sent_tokenize(x)
```

```python
df['sent_split']=df['doc'].apply(sentsplitter)
```

```python
df['word_split']=df['doc'].apply(wordsplitter)
```

```python
df.head()
```

```python
def counter(x):
    return len(x)
```

```python
df['count_speech_words']=df['word_split'].apply(counter)
df['count_speech_sent']=df['sent_split'].apply(counter)
```

```python
df.to_json("un_speeches_tokenised.json",orient='records',compression='infer')
```
